```python
import numpy as np

np.random.seed(42)

def tanh(x):
    return np.tanh(x)

x = np.array([0.6, -0.2])

input_size = 2
hidden_size = 2
output_size = 1

W1 = np.random.uniform(-0.5, 0.5, (hidden_size, input_size))
b1 = np.array([0.5, 0.5])

W2 = np.random.uniform(-0.5, 0.5, (output_size, hidden_size))
b2 = np.array([0.7])

z1 = np.dot(W1, x) + b1
a1 = tanh(z1)

z2 = np.dot(W2, a1) + b2
a2 = tanh(z2)

print("Output of the network:", a2[0])
```
